```python
#import library
from transformers import pipeline
```

```python
# Load the pre-trained summarization model
summarizer = pipeline("summarization")
```

```python
# Input text to be summarized
input_text = """
Natural language processing (NLP) is a subfield of artificial intelligence that focuses on the interaction between computers and humans through natural language. It involves tasks such as language generation, understanding, translation, and more. NLP has applications in various domains, including chatbots, sentiment analysis, language translation, and text summarization.
"""
```

```python
# Perform text summarization
summary = summarizer(input_text, max_length=40, min_length=30, do_sample=False)[0]['summary_text']
# Print the generated summary
```

```python
print(summary)
```

```python
from transformers import BartForConditionalGeneration, BartTokenizer

# Load pre-trained BART model and tokenizer
model_name = "facebook/bart-large-cnn"
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

```

```python
# Input text to be summarized
input_text = """
Natural language processing (NLP) is a subfield of artificial intelligence that focuses on the interaction between computers and humans through natural language. It involves tasks such as language generation, understanding, translation, and more. NLP has applications in various domains, including chatbots, sentiment analysis, language translation, and text summarization.
"""

# Tokenize input text
inputs = tokenizer.encode("summarize: " + input_text, return_tensors="pt", max_length=1024, truncation=True)

# Generate summary
summary_ids = model.generate(inputs, max_length=50, min_length=30, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

```

```python

# Print the generated summary
print(summary)
```

```python
from transformers import BartForConditionalGeneration, BartTokenizer

# Load pre-trained BART model and tokenizer
model_name = "facebook/bart-large-cnn"
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

# Input text to be summarized
input_text = """
Exercise in nature is very beneficial for both physical and mental health. In addition to the general benefits of exercise, such as improving cardiovascular health and increasing muscle strength, outdoor activities also provide an opportunity to experience the beauty and tranquillity of nature. It can increase feelings of well-being and reduce stress and anxiety. It is important to embrace the benefits of outdoor exercise and incorporate it into your regular routine.
"""

# Tokenize input text
inputs = tokenizer.encode(input_text, return_tensors="pt", max_length=1024, truncation=True)

# Generate summary
summary_ids = model.generate(inputs, max_length=150, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Print the generated summary
print(summary)
```
